```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data= pd.read_csv('used cars data.csv')
```

```python
data['price_diff']= data['Present_Price']- data['Selling_Price']
```

```python
data
```

```python
car_dataset = data.drop(['Car_Name'],axis=1)
```

```python
car_dataset['Current_year']= 2023
```

```python
car_dataset["Number_of_years"] = car_dataset['Current_year'] - data['Year']
```

```python
car_dataset
```

```python
car_dataset.drop(['Year'],axis=1)
```

```python
car_dataset.head()
```

```python
car_dataset = pd.get_dummies(car_dataset,drop_first=True)
car_dataset.head()
```

```python
correlation= car_dataset.corr()
plt.figure(figsize=(14,8))
sns.heatmap(correlation.corr(), annot=True,cmap='YlOrBr')
```

```python
x = car_dataset.iloc[:,1:]
y = car_dataset.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.ensemble import ExtraTreesRegressor
model = ExtraTreesRegressor()
model.fit(x,y)
```

```python
print(model.feature_importances_)
```

```python
feat_importances = pd.Series(model.feature_importances_, index=x.columns)
feat_importances.nlargest(5).plot(kind='bar')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_random = RandomForestRegressor()
```

```python
n_estimators=[int(x) for x in np.linspace(start = 100, stop=1200, num=12)]

max_features=['auto','sqrt']

max_depth=[int(x) for x in np.linspace(5,30, num=6)]

min_samples_split=[2,5,10,100]

min_samples_leaf = [1,2,5,10]
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
random_grid = {'n_estimators':n_estimators,
                "max_features":max_features,
                "max_depth":max_depth,
                "min_samples_split":min_samples_split,
                'min_samples_leaf':min_samples_leaf}
```

```python
rf = RandomForestRegressor()
```

```python
rf_random = RandomizedSearchCV(estimator=rf, param_distributions =random_grid,scoring='neg_mean_squared_error',n_iter=10,cv=5,verbose=2,random_state=42,n_jobs=1)
```

```python
rf_random.fit(x_train, y_train)
```

```python
rf_random.best_params_
```

```python
predictions= rf_random.predict(x_test)
predictions
```

```python
sns.displot(predictions)
```
